```python


def unzip_tar_files():
    path_of_zipped_files = rf"{os.getcwd()}\src\zipped"
    path_of_unzipped_files = rf"{os.getcwd()}\src\unzipped"
    for tar_file in os.listdir(path_of_zipped_files):
        print(f"Current file : {tar_file}")
        current_file = tarfile.open(path_of_zipped_files + "\\" + tar_file)
        current_file.extractall(path_of_unzipped_files)
        current_file.close()


unzip_tar_files()
```

```python
def crop_maharashtra(file_path):
    data = xr.open_dataset(file_path)
    df = data['VHI'].to_dataframe()
    df = df.reset_index(['HEIGHT', 'WIDTH'])

    dfx = pd.DataFrame()
    for index, row in df.loc[1378 * 10000 : 1655 * 10000].iterrows():
        if 7018 < row['WIDTH'] < 7247:
            dfx.at[row['HEIGHT'], row['WIDTH']] = row['VHI']
    
    return dfx
```

```python
def crop_dataset():
    path_of_unzipped_files = rf"{os.getcwd()}\src\unzipped"
    path_of_cropped_files = rf"{os.getcwd()}\src\cropped"
    
    for file in os.listdir(path_of_unzipped_files):
        cropped_dataframe = crop_maharashtra(path_of_unzipped_files + "/" + file)
        cropped_dataframe.to_csv(f"{path_of_cropped_files}/{file}.csv")
        

crop_dataset()
```

```python
def rename_files():
    path_of_unzipped_files = rf"{os.getcwd()}\src\unzipped"
    for file in os.listdir(path_of_unzipped_files):
        os.rename(f"{path_of_unzipped_files}/{file}", f"{path_of_unzipped_files}/{file[-13:-6]}")

rename_files()
```

```python
# 2000 - 2023

import warnings
warnings.filterwarnings('ignore') 


def crop_dataset():
    path_of_unzipped_files = rf"{os.getcwd()}\src\unzipped"
    path_of_cropped_files = rf"{os.getcwd()}\src\cropped"
    count = 0
    print("start time : 00:00")
    start = time.time()
    curr = start
    for file in os.listdir(path_of_unzipped_files):
        if file[0:2] == "20":
            cropped_dataframe = crop_maharashtra(path_of_unzipped_files + "/" + file)
            cropped_dataframe.to_csv(f"{path_of_cropped_files}/{file[:-3]}.csv")
            count += 1
            if count % 10 == 0:
                print("file : ", count, ", req time : ", time.time() - curr)
                curr = time.time()
    print("total req time : ", time.time() - start)


crop_dataset()
```
